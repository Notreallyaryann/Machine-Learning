```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"/content/Ice_cream selling data.csv")
```

```python
df.head()
```

```python
x=df.iloc[:,:-1]
y=df["Ice Cream Sales (units)"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt=DecisionTreeRegressor(criterion='poisson',splitter='best',max_depth=5)
dt.fit(x_train,y_train)
```

```python
dt.score(x_test,y_test)*100,dt.score(x_train,y_train)*100
```

```python
from sklearn.model_selection import GridSearchCV,RandomizedSearchCV
```

```python
df = {
    'criterion': ['squared_error',"friedman_mse","absolute_error","poisson"],
    'splitter': ['best',"random"],
    'max_depth':[i for i in range(2,20)]
}
```

```python
rd=RandomizedSearchCV(DecisionTreeRegressor(),param_distributions=df,n_iter=20)
rd.fit(x_train,y_train)
```

```python
rd.best_params_
```

```python
rd.best_score_
```

```python
gd=GridSearchCV(DecisionTreeRegressor(),param_grid=df)
gd.fit(x_train,y_train)
```

```python
gd.best_params_
```

```python
gd.best_score_
```
