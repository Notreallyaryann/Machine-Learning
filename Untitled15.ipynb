Apriori Algo(Association Rule Learning)

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r"/content/groceries - groceries.csv")
```

```python
df.head(3)
```

```python
new_df=df.drop(columns=["Item(s)"])
```

```python
new_df.head()
```

```python
new_df.shape
```

```python
market=[]
for i in range(0,new_df.shape[0]):
  cus=[]
  for j in new_df.columns:
    if type(new_df[j][i])==str:
      cus.append(new_df[j][i])
  market.append(cus)
```

```python
l=[]
for i in market:
  for j in i:
    l.append(j)
```

```python
## most freq buy product
import collections
```

```python
p=collections.Counter(l)
```

```python
p.keys()
```

```python
p.values()
```

```python
d={"Item Name":p.keys(),"values":p.values()}
```

```python
pd.DataFrame(d).sort_values(by=["values"],ascending=False)
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
tr=TransactionEncoder()
tr.fit(market)
```

```python
df1=pd.DataFrame(tr.transform(market),columns=tr.columns_)
```

```python
from mlxtend.frequent_patterns import apriori
```

```python
apriori(df1,min_support=0.05,use_colnames=True,max_len=3).sort_values(by=["support"])
```
