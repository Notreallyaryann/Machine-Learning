Support Vector Machine

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
df=pd.read_csv(r"/content/placement.csv")
```

```python
df.head()
```

```python
df=df.drop(columns=["Unnamed: 0"])
```

```python
df.head()
```

```python
sns.scatterplot(x="cgpa",y="iq",data=df,hue="placement")
plt.show()
```

```python
x=df.iloc[:, [0, -1]]
y=df["placement"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.svm import SVC
```

```python
#kernels->>("linear","poly","rbf","sigmoid",callable)
sv=SVC(kernel="poly")
sv.fit(x_train,y_train)
```

```python

sv.score(x_test,y_test)*100
```

```python
sv.score(x_train,y_train)*100
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=sv)
plt.show()
```

Support vector Machine Regression

```python
df=pd.read_csv(r"/content/placement-2.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
sns.scatterplot(x="cgpa",y="package",data=df)
plt.show()
```

```python
x=df[["cgpa"]]
y=df["package"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.svm import SVR
```

```python
sv=SVR(kernel="linear")
sv.fit(x_train,y_train)
```

```python
sv.score(x_test,y_test)*100
```

```python
sv.score(x_train,y_train)*100
```

```python
sns.scatterplot(x="cgpa",y="package",data=df)
plt.plot(df["cgpa"],sv.predict(x),color="red")
plt.show()
```
