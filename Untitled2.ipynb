Outliers

```python
sns.boxplot(x="CoapplicantIncome",data=df)
```

```python
sns.displot(df["ApplicantIncome"])
plt.show()
```

```python
# Calculate Q1 and Q3
q1 = df["CoapplicantIncome"].quantile(0.25)
q3 = df["CoapplicantIncome"].quantile(0.75)

# Calculate IQR
IQR = q3 - q1


min_range = q1 - (1.5 * IQR)
max_range = q3 + (1.5 * IQR)


print("Min Range:", min_range, "Max Range:", max_range)


df_filtered = df[(df["CoapplicantIncome"] >= min_range) & (df["CoapplicantIncome"] <= max_range)]
```

```python
df_no_outliers = df[(df["CoapplicantIncome"] >= min_range) & (df["CoapplicantIncome"] <= max_range)]
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x=df_no_outliers["CoapplicantIncome"])
plt.title("Boxplot of Coapplicant Income (Outliers Removed)")
plt.xlabel("Coapplicant Income")
plt.show()
```

```python
#used to standardize (scale) numerical features by removing the mean and scaling to unit variance
from sklearn.preprocessing import StandardScaler
```

```python
df.describe()
```

```python

#plt.subplot(nrows, ncols, index)

plt.subplot(1,2,1)
plt.title("Before")
sns.distplot(df["ApplicantIncome_ss"])


plt.subplot(1,2,2)
plt.title("After")
sns.distplot(df["ApplicantIncome"])


plt.show()
```
